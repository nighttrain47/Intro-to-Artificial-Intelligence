## Cơ Sở Lý Thuyết

### 1. Tìm Kiếm Mù
Tìm kiếm mù (Uninformed Search) là các thuật toán tìm kiếm không sử dụng thông tin heuristic, duyệt qua không gian trạng thái theo cách có hệ thống, dựa trên cấu trúc bài toán.

### 2. Tìm Kiếm theo Chiều Rộng (BFS)
- **Nguyên lý**: Khám phá tất cả các nút ở mức hiện tại trước khi chuyển sang mức tiếp theo.
- **Cấu trúc dữ liệu**: Hàng đợi (Queue).
- **Đặc điểm**:
  - Trong đồ thị không trọng số, BFS đảm bảo tìm đường đi ngắn nhất (theo số cạnh).
  - Trong đồ thị có trọng số, BFS không đảm bảo đường đi có tổng trọng số nhỏ nhất, vì nó không xem xét trọng số của các cạnh.

### 3. Tìm Kiếm theo Chiều Sâu (DFS)
- **Nguyên lý**: Khám phá một nhánh đến tận cùng trước khi quay lại thử nhánh khác.
- **Cấu trúc dữ liệu**: Ngăn xếp (Stack) hoặc đệ quy.
- **Đặc điểm**:
  - DFS không đảm bảo đường đi ngắn nhất trong cả đồ thị không trọng số và có trọng số.
  - Kết quả phụ thuộc vào thứ tự duyệt các nút kề.

### 4. Đồ Thị Không Trọng Số và Có Trọng Số
- **Đồ thị không trọng số**: Các cạnh có trọng số ngầm định bằng 1. BFS tối ưu theo số cạnh.
- **Đồ thị có trọng số**: Các cạnh có trọng số khác nhau (ví dụ: chi phí, khoảng cách). BFS và DFS không được thiết kế để tối ưu hóa tổng trọng số, dẫn đến kết quả không tối ưu.

### 5. Bài Toán 
Cho các đồ thị vô hướng (không trọng số và có trọng số) biểu diễn mạng lưới hoặc mê cung. Nhiệm vụ là tìm đường đi từ nút bắt đầu (**S**) đến nút đích (**G** hoặc **H**) bằng BFS và DFS, thông qua minh họa thủ công và lập trình với chú thích chi tiết.

### Phần 3: Triển Khai Mã Python

#### BFS trên Đồ Thị Không Trọng Số
Chạy ô mã dưới đây để kiểm chứng BFS trên **Đồ thị mẫu 1**.

````python
from collections import deque

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 1 (không trọng số)
graph1 = {
    'S': ['A', 'B', 'C'],
    'A': ['S', 'D'],
    'B': ['S', 'E'],
    'C': ['S'],
    'D': ['A', 'F'],
    'E': ['B', 'G'],
    'F': ['D'],
    'G': ['E']
}

# Chạy BFS và in kết quả
path = bfs(graph1, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 1:", path)
````

#### DFS trên Đồ Thị Không Trọng Số
Chạy ô mã dưới đây để kiểm chứng DFS trên **Đồ thị mẫu 1**.

````python
def dfs(graph, start, goal, visited=None, path=None):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path
    
    # Duyệt qua các nút kề của nút hiện tại
    for neighbor in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi mới
            new_path = dfs(graph, neighbor, goal, visited, path + [neighbor])
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 1 (không trọng số)
graph1 = {
    'S': ['A', 'B', 'C'],
    'A': ['S', 'D'],
    'B': ['S', 'E'],
    'C': ['S'],
    'D': ['A', 'F'],
    'E': ['B', 'G'],
    'F': ['D'],
    'G': ['E']
}

# Chạy DFS và in kết quả
path = dfs(graph1, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 1:", path)
````

#### BFS trên Đồ Thị Có Trọng Số
Chạy ô mã dưới đây để kiểm chứng BFS trên **Đồ thị mẫu 5**.

````python
from collections import deque

def bfs_weighted(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu, đường đi, và tổng trọng số
    queue = deque([(start, [start], 0)])
    # Khởi tạo tập hợp các nút đã thăm
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút, đường đi, và tổng trọng số từ đầu hàng đợi
        (node, path, total_weight) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path, total_weight
        # Duyệt qua các nút kề và trọng số tương ứng
        for neighbor, weight in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề, đường đi mới, và tổng trọng số cập nhật vào hàng đợi
                queue.append((neighbor, path + [neighbor], total_weight + weight))
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 5 (có trọng số)
graph5 = {
    'S': [('A', 1), ('B', 2), ('C', 3)],
    'A': [('S', 1), ('D', 4), ('G', 3)],
    'B': [('S', 2), ('G', 5)],
    'C': [('S', 3)],
    'D': [('A', 4), ('G', 6)],
    'G': [('B', 5), ('D', 6), ('A', 3)]
}

# Chạy BFS và in kết quả
path, weight = bfs_weighted(graph5, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 5:", path, "Tổng trọng số:", weight)
````

#### DFS trên Đồ Thị Có Trọng Số
Chạy ô mã dưới đây để kiểm chứng DFS trên **Đồ thị mẫu 5**.

````python
def dfs_weighted(graph, start, goal, visited=None, path=None, total_weight=0):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path, total_weight
    
    # Duyệt qua các nút kề và trọng số tương ứng
    for neighbor, weight in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi và tổng trọng số mới
            new_path, new_weight = dfs_weighted(graph, neighbor, goal, visited, path + [neighbor], total_weight + weight)
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path, new_weight
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 5 (có trọng số)
graph5 = {
    'S': [('A', 1), ('B', 2), ('C', 3)],
    'A': [('S', 1), ('D', 4), ('G', 3)],
    'B': [('S', 2), ('G', 5)],
    'C': [('S', 3)],
    'D': [('A', 4), ('G', 6)],
    'G': [('B', 5), ('D', 6), ('A', 3)]
}

# Chạy DFS và in kết quả
path, weight = dfs_weighted(graph5, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 5:", path, "Tổng trọng số:", weight)
````

## Bài Tập lập trình

#### Bài Tập Lập Trình: BFS trên Đồ Thị Mẫu 2
Chạy BFS trên **Đồ thị mẫu 2** với chú thích chi tiết.

````python
from collections import deque

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 2 (không trọng số)
graph2 = {
    'S': ['A', 'B'],
    'A': ['S', 'B', 'C'],
    'B': ['S', 'D'],
    'C': ['A', 'D'],
    'D': ['B', 'G'],
}

# Chạy BFS và in kết quả
path = bfs(graph2, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 2:", path)
````

#### Bài Tập Lập Trình: DFS trên Đồ Thị Mẫu 2
Chạy DFS trên **Đồ thị mẫu 2** với chú thích chi tiết.

````python
def dfs(graph, start, goal, visited=None, path=None):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path
    
    # Duyệt qua các nút kề của nút hiện tại
    for neighbor in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi mới
            new_path = dfs(graph, neighbor, goal, visited, path + [neighbor])
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 2 (không trọng số)
graph2 = {
    'S': ['A', 'B'],
    'A': ['S', 'B', 'C'],
    'B': ['S', 'D'],
    'C': ['A', 'D'],
    'D': ['B', 'G'],
}

# Chạy DFS và in kết quả
path = dfs(graph2, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 2:", path)
````

````python
#BFS đồ thị mẫu 6

from collections import deque

def bfs_weighted(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu, đường đi, và tổng trọng số
    # Mỗi phần tử trong hàng đợi là một tuple: (node hiện tại, danh sách đường đi, tổng trọng số của đường đi đó)
    queue = deque([(start, [start], 0)])
    # Khởi tạo tập hợp các nút đã thăm
    # Trong BFS cho đường đi ngắn nhất (không trọng số), chúng ta thường chỉ thêm vào visited khi lần đầu khám phá.
    # Tuy nhiên, trong trường hợp này, với graph có trọng số, BFS không đảm bảo tìm ra đường đi ngắn nhất về trọng số,
    # mà là đường đi ngắn nhất về số cạnh. Để tìm đường đi ngắn nhất về trọng số, cần dùng Dijkstra.
    # Với mục đích tìm BẤT KỲ đường đi từ start đến goal bằng BFS, cách này vẫn hoạt động.
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút, đường đi, và tổng trọng số từ đầu hàng đợi
        (node, path, total_weight) = queue.popleft()
        
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path, total_weight
        
        # Duyệt qua các nút kề và trọng số tương ứng
        # Sử dụng .items() để lấy cả khóa (neighbor) và giá trị (weight) từ từ điển con graph[node]
        for neighbor, weight in graph[node].items():
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm để tránh lặp lại hoặc chu trình vô hạn
                visited.add(neighbor)
                # Thêm nút kề, đường đi mới, và tổng trọng số cập nhật vào hàng đợi
                queue.append((neighbor, path + [neighbor], total_weight + weight))
    
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 6 (có trọng số)
graph6 = {
    'S': {'A': 2, 'C': 5},
    'A': {'B': 3, 'D': 4},
    'B': {'E': 6},
    'C': {'D': 7, 'F': 9},
    'D': {'E': 8},
    'E': {'H': 10},
    'F': {'G': 12},
    'G': {'H': 15},
    'H': {} 
}

# Chạy BFS và in kết quả
path, weight = bfs_weighted(graph6, 'S', 'H')
print("Đường đi BFS trên Đồ thị mẫu 6:", path, "Tổng trọng số:", weight)
````

````python
#DFS đồ thị mẫu 6

def dfs_weighted(graph, start, goal, visited=None, path=None, total_weight=0):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    # visited phải là một tập hợp mới cho mỗi lần gọi đệ quy để cho phép khám phá các nhánh khác nhau
    # nhưng để tìm MỘT đường đi đầu tiên mà DFS tìm thấy, chúng ta cần một `visited` chung cho toàn bộ quá trình đệ quy ban đầu.
    # Tuy nhiên, cách triển khai DFS này sẽ tìm đường đi **đầu tiên** mà nó khám phá (dựa trên thứ tự duyệt neighbors),
    # chứ không phải đường đi ngắn nhất về trọng số hay số cạnh.
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path, total_weight
    
    # Duyệt qua các nút kề và trọng số tương ứng
    # Sửa lỗi: dùng .items() để lấy cả khóa (neighbor) và giá trị (weight) từ từ điển con
    for neighbor, weight in graph[start].items():
        # Chỉ đi vào nút kề nếu nó chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi và tổng trọng số mới
            # Cần truyền cùng một tập `visited` để tránh lặp lại các nút trong cùng một đường đi khám phá
            new_path, new_weight = dfs_weighted(graph, neighbor, goal, visited, path + [neighbor], total_weight + weight)
            
            # Nếu tìm thấy đường đi từ nút kề đến đích, trả về kết quả
            if new_path:
                return new_path, new_weight
    
    # Trả về None và 0 nếu không tìm thấy đường đi từ nút hiện tại
    return None, 0

# Định nghĩa Đồ thị mẫu 6 (có trọng số)
graph6 = {
    'S': {'A': 2, 'C': 5},
    'A': {'B': 3, 'D': 4},
    'B': {'E': 6},
    'C': {'D': 7, 'F': 9},
    'D': {'E': 8},
    'E': {'H': 10},
    'F': {'G': 12},
    'G': {'H': 15},
    'H': {} 
}

# Chạy DFS và in kết quả
path, weight = dfs_weighted(graph6, 'S', 'H')
print("Đường đi DFS trên Đồ thị mẫu 6:", path, "Tổng trọng số:", weight)
````

````python
#BFS đồ thị mẫu 7

from collections import deque

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 7 (không trọng số)
graph7 = graph7 = {
    'S': ['A', 'D', 'E'],
    'A': ['B', 'D', 'E'],
    'B': ['C', 'E', 'F'],
    'C': ['F', 'G'],
    'D': ['E'],
    'E': ['F', 'H'],
    'F': ['G', 'H'],
    'G': ['H'],
    'H': [] }

# Chạy BFS và in kết quả
path = bfs(graph7, 'S', 'H')
print("Đường đi BFS trên Đồ thị mẫu 7:", path)
````

````python
# DFS đồ thị mẫu 7

def dfs(graph, start, goal, visited=None, path=None):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path
    
    # Duyệt qua các nút kề của nút hiện tại
    for neighbor in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi mới
            new_path = dfs(graph, neighbor, goal, visited, path + [neighbor])
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 7 (không trọng số)
graph7 = graph7 = {
    'S': ['A', 'D', 'E'],
    'A': ['B', 'D', 'E'],
    'B': ['C', 'E', 'F'],
    'C': ['F', 'G'],
    'D': ['E'],
    'E': ['F', 'H'],
    'F': ['G', 'H'],
    'G': ['H'],
    'H': [] }

# Chạy DFS và in kết quả
path = dfs(graph7, 'S', 'H')
print("Đường đi DFS trên Đồ thị mẫu 7:", path)
````

````python
#Sửa lỗi BFS để lưu tất cả các đường đi từ start đến goal

from collections import deque

def bfs_all_paths(graph, start, goal):
    # Hàng đợi sẽ lưu các đường đi tìm được, không chỉ các nút đơn lẻ
    queue = deque([[start]]) 
    # Danh sách để lưu trữ tất cả các đường đi tìm thấy
    all_paths = []

    while queue:
        path = queue.popleft()
        current_node = path[-1] # Lấy nút cuối cùng của đường đi hiện tại

        # Nếu nút hiện tại là đích, thì đây là một đường đi hợp lệ
        if current_node == goal:
            all_paths.append(path)
            # Không return ngay lập tức vì chúng ta muốn tìm TẤT CẢ các đường đi
            continue # Tiếp tục vòng lặp để khám phá các đường đi khác

        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[current_node]:
            # Quan trọng: Kiểm tra xem nút kề đã CÓ TRONG ĐƯỜNG ĐI HIỆN TẠI chưa để tránh chu trình vô hạn trong MỘT ĐƯỜNG ĐI CỤ THỂ.
            # Chúng ta không dùng global visited set nữa vì muốn khám phá lại các nút qua các đường đi khác nhau.
            if neighbor not in path: 
                new_path = list(path) # Tạo một bản sao của đường đi hiện tại
                new_path.append(neighbor) # Thêm nút kề vào bản sao đường đi mới
                queue.append(new_path) # Thêm đường đi mới vào hàng đợi
                
    return all_paths # Trả về danh sách tất cả các đường đi

# Định nghĩa Đồ thị mẫu 7 (không trọng số)
graph7 = {
    'S': ['A', 'D', 'E'],
    'A': ['B', 'D', 'E'],
    'B': ['C', 'E', 'F'],
    'C': ['F', 'G'],
    'D': ['E'],
    'E': ['F', 'H'],
    'F': ['G', 'H'],
    'G': ['H'],
    'H': [] 
}

# Chạy BFS để tìm tất cả các đường đi và in kết quả
all_found_paths = bfs_all_paths(graph7, 'S', 'H')
print("Tất cả các đường đi BFS từ S đến H trên Đồ thị mẫu 7:")
if all_found_paths:
    for p in all_found_paths:
        print(p)
else:
    print("Không tìm thấy đường đi nào từ S đến H.")
````

````python
# Tạo đồ thị 

import time
from collections import deque

# --- Hàm BFS cho đồ thị có trọng số ---
def bfs_weighted(graph, start, goal):
    """
    Thực hiện tìm kiếm theo chiều rộng (BFS) trên đồ thị có trọng số.
    Thuật toán này tìm ra đường đi ngắn nhất về SỐ CẠNH (số bước nhảy) từ start đến goal.
    NÓ KHÔNG ĐẢM BẢO đường đi ngắn nhất về TỔNG TRỌNG SỐ.

    Args:
        graph (dict): Biểu diễn đồ thị dưới dạng từ điển lồng từ điển (node: {neighbor: weight}).
        start (str): Nút bắt đầu.
        goal (str): Nút đích.

    Returns:
        tuple: (list) Đường đi tìm được, (int) Tổng trọng số của đường đi đó.
               Trả về (None, 0) nếu không tìm thấy đường đi.
    """
    queue = deque([(start, [start], 0)]) # Lưu trữ: (nút hiện tại, đường đi tới nút đó, tổng trọng số của đường đi)
    visited = set([start]) # Tập hợp các nút đã thăm để tránh lặp trong quá trình BFS

    while queue:
        node, path, total_weight = queue.popleft() # Lấy đường đi ngắn nhất về số cạnh hiện tại từ hàng đợi

        if node == goal:
            return path, total_weight # Đã tìm thấy đích, trả về đường đi

        # Duyệt qua tất cả các nút kề và trọng số tương ứng
        for neighbor, weight in graph[node].items(): # Dùng .items() để lấy cả khóa (neighbor) và giá trị (weight)
            if neighbor not in visited: # Nếu nút kề chưa được thăm
                visited.add(neighbor) # Thêm nút kề vào tập đã thăm
                # Tạo đường đi mới và cập nhật tổng trọng số, sau đó thêm vào hàng đợi
                queue.append((neighbor, path + [neighbor], total_weight + weight))
    
    return None, 0 # Không tìm thấy đường đi đến đích

# --- Hàm DFS cho đồ thị có trọng số ---
def dfs_weighted(graph, start, goal, visited=None, path=None, total_weight=0):
    """
    Thực hiện tìm kiếm theo chiều sâu (DFS) trên đồ thị có trọng số.
    Thuật toán này tìm MỘT đường đi đầu tiên từ start đến goal theo chiều sâu.
    NÓ KHÔNG ĐẢM BẢO đường đi ngắn nhất về trọng số hay số cạnh.

    Args:
        graph (dict): Biểu diễn đồ thị dưới dạng từ điển lồng từ điển (node: {neighbor: weight}).
        start (str): Nút bắt đầu.
        goal (str): Nút đích.
        visited (set, optional): Tập hợp các nút đã thăm trong lần gọi đệ quy hiện tại. Mặc định là None.
        path (list, optional): Đường đi hiện tại đến nút start. Mặc định là None.
        total_weight (int, optional): Tổng trọng số của đường đi hiện tại. Mặc định là 0.

    Returns:
        tuple: (list) Đường đi tìm được, (int) Tổng trọng số của đường đi đó.
               Trả về (None, 0) nếu không tìm thấy đường đi từ nhánh hiện tại.
    """
    # Khởi tạo tập đã thăm và đường đi nếu chưa có (chỉ lần gọi ban đầu)
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    visited.add(start) # Thêm nút hiện tại vào tập đã thăm
    
    if start == goal:
        return path, total_weight # Đã tìm thấy đích, trả về đường đi và tổng trọng số
    
    # Duyệt qua các nút kề và trọng số tương ứng
    for neighbor, weight in graph[start].items(): # Dùng .items() để lấy cả khóa (neighbor) và giá trị (weight)
        if neighbor not in visited: # Chỉ đi vào nút kề nếu nó chưa được thăm
            # Gọi đệ quy DFS trên nút kề với đường đi và tổng trọng số mới
            # Truyền cùng một tập `visited` để tránh lặp lại các nút trong cùng một đường đi
            new_path, new_weight = dfs_weighted(graph, neighbor, goal, visited, path + [neighbor], total_weight + weight)
            
            if new_path: # Nếu tìm thấy đường đi từ nhánh nút kề đến đích, trả về kết quả
                return new_path, new_weight
    
    return None, 0 # Không tìm thấy đường đi từ nút hiện tại hoặc nhánh của nó

graph_complex = {
    'S': {'A': 8, 'B': 12},
    'A': {'C': 5, 'D': 15},
    'B': {'D': 7, 'E': 10},
    'C': {'F': 9, 'G': 4},
    'D': {'G': 6, 'H': 11},
    'E': {'H': 13},
    'F': {'I': 3},
    'G': {'I': 18, 'J': 2},
    'H': {'J': 7},
    'I': {'K': 10},
    'J': {'K': 5},
    'K': {} 
}

start_node = 'S'
goal_node = 'K'

print(f"--- Tìm đường đi từ {start_node} đến {goal_node} trên Đồ thị ---")

# --- Chạy và đo thời gian cho BFS ---
start_time_bfs = time.perf_counter()
path_bfs, weight_bfs = bfs_weighted(graph_complex, start_node, goal_node)
end_time_bfs = time.perf_counter()
print(f"BFS: Đường đi = {path_bfs}")
print(f"     Tổng trọng số = {weight_bfs}")
print(f"     Thời gian chạy = {end_time_bfs - start_time_bfs:.6f} giây")

print("\n" + "="*50 + "\n") # Dải phân cách

# --- Chạy và đo thời gian cho DFS ---
start_time_dfs = time.perf_counter()
path_dfs, weight_dfs = dfs_weighted(graph_complex, start_node, goal_node)
end_time_dfs = time.perf_counter()
print(f"DFS: Đường đi = {path_dfs}")
print(f"     Tổng trọng số = {weight_dfs}")
print(f"     Thời gian chạy = {end_time_dfs - start_time_dfs:.6f} giây")
````
